```python
import pandas as pd
import json

# Load depression dataset
depression_df = pd.read_csv("student_depression_dataset.csv")

# Load and convert performance dataset
with open("data.json", "r") as f:
    performance_data = json.load(f)
performance_df = pd.json_normalize(performance_data)

# Save for later steps
performance_df.to_csv("student_performance.csv", index=False)

# Preview
print(depression_df.shape)
print(performance_df.shape)
depression_df.head()
```
