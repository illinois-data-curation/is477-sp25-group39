```python
import pandas as pd

# Load raw data
depression_df = pd.read_csv("student_depression_dataset.csv")
performance_df = pd.read_csv("student_performance.csv")

# Standardize column names
depression_df.columns = depression_df.columns.str.lower().str.replace(" ", "_")
performance_df.columns = performance_df.columns.str.lower().str.replace(" ", "_")

# Handle missing values
depression_df.fillna(depression_df.median(numeric_only=True), inplace=True)
performance_df.fillna(performance_df.median(numeric_only=True), inplace=True)

# Save cleaned versions
depression_df.to_csv("cleaned_depression.csv", index=False)
performance_df.to_csv("cleaned_performance.csv", index=False)
```
